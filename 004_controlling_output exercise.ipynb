````python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

client = Anthropic() 
model = "claude-sonnet-4-5"
````

````python

# Helpers
def add_user_message(messages, text):
  user_message = {
    "role": "user", 
    "content": text
  }
  messages.append(user_message)

def add_assistant_message(messages, text):
  assistant_message = {
    "role": "assistant",
    "content": text
  }
  messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0, stop_sequences=[]):
  params= {
    "model": model,
    "max_tokens": 1000,
    "messages": messages,
    "temperature": temperature,
    "stop_sequences": stop_sequences
  }

  if system:
    params["system"] = system

  message = client.messages.create(**params)
  return message.content[0].text
````

````python
messages = []

prompt = """
Generate three different sample AWS CLI commands. Each should be very short.
"""

add_user_message(messages, prompt)
# using "```bash" turns out makes it that we can get all strings inside ```bash [here] ```
# kinda different with JSON where it only have one whole json string,
# which not really explained in previous course, turns out it can get multiple instead of only 1
add_assistant_message(
  messages, 
  # not sure what the difference with using only ```bash or with using begining text
  # possibly it was to avoid additional text given in rseponse, makign it more reliable
  # makign assistan could guide claude dramatically
  "```bash"
  # "Here are all three commands : \n```bash"
  # ":"
  # "{"
  # "```"
)
# add_assistant_message(messages, "}")
text = chat(
  messages, 
  stop_sequences=[
    "```"
    # "{"
  ]
)
text.strip()
````
