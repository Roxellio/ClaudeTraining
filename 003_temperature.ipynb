```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

client = Anthropic() 
model = "claude-sonnet-4-6"
```

```python

# Helpers
def add_user_message(messages, text):
  user_message = {
    "role": "user", 
    "content": text
  }
  messages.append(user_message)

def add_assistant_message(messages, text):
  assistant_message = {
    "role": "user",
    "content": text
  }
  messages.append(assistant_message)

# temperature behaves to affect on how different the answer would be
# the lower it is, the more similar the answer be. Making it rather similar in nature
# such as Alzheimer cracker keeps comign out if set to 0
# the higher it is, the more different in nature
# suchas from Alzhermier safecracker would hanges to nurses, etc
# basically less creativity/more deterministic, lower temperature. More creativity, needs more temperature
# values is decimal from 0 to 1
# low temp (0.0 - 0.3) -> factual stuff, assitance in codes/math/data/etc
# med temp (0.4 - 0.7) -> summary, edu content, problem solving, creative writing with constraint
# hi temp (0.8 - 1.0) -> brainstorming, creative writing, marketing, joke generation
def chat(messages, system=None, temperature=1.0):
  params= {
    "model": model,
    "max_tokens": 1000,
    "messages": messages,
    "temperature": temperature
  }

  if system:
    params["system"] = system

  message = client.messages.create(**params)
  return message.content[0].text
```

```python
messages = []

add_user_message(messages, "Generate a one sentence movie idea")
answer = chat(messages, temperature=1.0)
answer
```
