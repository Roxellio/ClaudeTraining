```python
# Load env variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API Client
from anthropic import Anthropic

client = Anthropic()
# do not use old model, it might cause errors due to deprecated stuff
# model = "claude-sonnnet-4-0"  # or "claude-sonnnet-4-0-20240926" for the latest version
# model = "claude-sonnnet-4-0-20240926"
model = "claude-haiku-4-5-20251001"
```

```python
# Make a request
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)

# Message result
message
```
